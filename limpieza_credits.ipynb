# Exploracion y transformacion de la columna 'cast'

```python
credits['cast']
dato1 = credits['cast'][0]
type(dato1)
```

```python
# listas de lo que parecen ser diccionarios, y listas vacías.
# procedemos a la transformación de texto a diccionario
import ast
import json

def convertir_a_dict(valor):
    datos_list = ast.literal_eval(valor)
    datos_str_list = [json.dumps(d) for d in datos_list]
    datos_dict_list = [json.loads(d) for d in datos_str_list]
    return datos_dict_list

# Aplicar la función a toda la columna 'cast' agregando una nueva columna al DF llamada 'cast_dict'
credits['cast_dict'] = credits['cast'].apply(convertir_a_dict)
```

```python

dato2 = credits['cast_dict'][0]
dato2
```

```python
valores_name = credits['cast_dict'].apply(lambda x: [d['order'] for d in x] if isinstance(x, list) else None)
valores_name
```

# Exploración y transformación de la columna 'crew'

```python
credits['crew']
dato1 = credits['crew'][0]
dato1
```

```python
# listas de lo que parecen ser diccionarios, y listas vacías.
# procedemos a la transformación de texto a diccionario
import ast
import json

def convertir_a_dict(valor):
    datos_list = ast.literal_eval(valor)
    datos_str_list = [json.dumps(d) for d in datos_list]
    datos_dict_list = [json.loads(d) for d in datos_str_list]
    return datos_dict_list

# Aplicar la función a toda la columna 'cast' agregando una nueva columna al DF llamada 'cast_dict'
credits['crew_dict'] = credits['crew'].apply(convertir_a_dict)
```

```python
dato2 = credits['crew_dict'][0]
dato2
```

```python
credits.info()
```

```python
valores_name = credits['crew_dict'].apply(lambda x: [d['name'] for d in x] if isinstance(x, list) else None)
valores_job = credits['crew_dict'].apply(lambda x: [d['job'] for d in x] if isinstance(x, list) else None)
print(valores_name)
print(valores_job)
```

```python
def obtener_nombres_y_trabajos(diccionarios):
    if isinstance(diccionarios, list):
        nombres = [d['name'] for d in diccionarios]
        trabajos = [d['job'] for d in diccionarios]
        return list(zip(nombres, trabajos))
    else:
        return None

valores = credits['crew_dict'].apply(obtener_nombres_y_trabajos)
```
